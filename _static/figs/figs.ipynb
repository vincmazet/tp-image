# Format images to be used in the book

```python
import numpy as np
import skimage.io as io
import skimage.color as color
from skimage.transform import resize, rescale
import matplotlib.pyplot as plt
```

```python
def format_image(filename, crop_origin, crop_size, size, newname, show=False):
    
    # Load image
    img = io.imread(f"../src/{filename}")
    print(f"Image of size : {img.shape}")
    
    # Color or grayscale image?
    color = (len(img.shape)==3)
    
    # Crop image
    r, c = crop_origin
    if color:
        img = img[ r:(r+crop_size) , c:(c+crop_size) , : ]
    else:
        img = img[ r:(r+crop_size) , c:(c+crop_size)]
    
    #Resize image
    if color:
        size = np.concatenate((size, [3]))
    print(size)
    img = resize(img, size, anti_aliasing=True)
    
    # Display image
    if show:
        plt.figure(figsize=(8,8))
        plt.imshow(img, cmap='gray')
        plt.show()
    
    # Save image
    io.imsave(newname, img)

    return img
```

```python
img = np.zeros((128, 128))

m, n = np.meshgrid(range(128),range(128))

spots = [
    [ 40, 100, 3, 1.00],
    [ 83,  65, 3, 0.10],
    [ 17,  32, 3, 0.10],
    [ 34,  79, 10, 0.05],
    [ 68,  12, 2, 0.2],
]
for spot in spots:
    x, y, sig, a = spot
    img += a * np.exp(-(m-x)**2/2/sig**2-(n-y)**2/2/sig**2)

# Affichage
plt.figure(figsize=(8, 8))
plt.imshow(img, cmap="gray")
plt.show()
io.imsave("spots.png", img)
```
